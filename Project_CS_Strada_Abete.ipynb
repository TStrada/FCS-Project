```python
import pandas as pd
```

```python
books = pd.read_csv('https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Books.csv.gz', encoding = 'latin-1', delimiter = ';', low_memory = False)
```

```python
book_ratings = pd.read_csv('https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Book-Ratings.csv.gz', encoding = 'latin-1', delimiter = ';', low_memory = False)
```

```python
users = pd.read_csv('https://github.com/gdv/foundationsCS/raw/master/progetti/2021/Users.csv.gz', encoding = 'latin-1', delimiter = ';', low_memory = False)
```

```python
goodbooks_rating = pd.read_csv('https://github.com/gdv/foundationsCS/raw/master/progetti/2021/goodbooks-ratings.csv.gz', encoding = 'latin-1')
```

```python
goodbooks = pd.read_csv('https://github.com/gdv/foundationsCS/raw/master/progetti/2021/goodbooks.csv.gz', encoding = 'latin-1')
```

**1) Normalize the location field of Users dataset, splitting into city, region, country.**

```python
location = users['Location'].str.split(', ', expand = True, n = 2).rename(columns = {0: 'City', 1: 'Region', 2: 'Country'})
location_split = pd.concat([users, location], axis = 1)
location_split
```

**2) For each book in the Books dataset, compute its average rating.**

```python
#Unione books e book_ratings. La richiameremo diverse volte
books_join_ratings = pd.merge(books, book_ratings, on = 'ISBN')
```

```python
rating_avg= books_join_ratings.groupby('ISBN', as_index = False).mean('Book-Rating').rename(columns = {'Book-Rating': 
                                                                                                       'Book_rating_avg'})
#Uniamo a books per avere tutte le informazioni disponibilin per ciascun libro
books_ratings_avg = pd.merge(rating_avg, books, on = 'ISBN')
books_ratings_avg.head(15)[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Book_rating_avg']]
```

**3) For each book in the GoodBooks dataset, compute its average rating.**

```python
goodbooks_join_ratings = pd.merge(goodbooks, goodbooks_rating, on = 'book_id') 
```

```python
avg_rating = goodbooks_join_ratings.groupby('book_id', as_index = False).mean('rating')
goodbooks_avg_rating = pd.merge(goodbooks, avg_rating, on = 'book_id').rename(columns={'rating': 'avg_rating'})
goodbooks_avg_rating[['book_id', 'isbn', 'authors', 'original_title', 'avg_rating']]
```

**4) Merge together all rows sharing the same book title, author and publisher. We will call the resulting datset merged books. The books that have not been merged together will not appear in merged books.**

```python
merged_books = books.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index = False).count().rename(columns={'ISBN':
                                                                                                                   'Count'})
#Consideriamo solo i libri che sono stati uniti ad almeno un altro libro
merged_books = merged_books[merged_books['Count'] > 1][['Book-Title', 'Book-Author', 'Publisher', 'Count']]
merged_books
```

**5) For each book in merged books compute its average rating. The average is computed considering all books in books that have been merged.**

```python
#Visualizziamo per ogni libro il corrispondente ISBN e book-rating
merged_books_ratings = pd.merge(merged_books, books_join_ratings, on = ['Book-Title', 'Book-Author', 'Publisher'])
merged_books_avg_rating = merged_books_ratings.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index = False).mean('Book-rating')
merged_books_avg_rating.rename(columns={'Book-Rating': 'Avg-Rating'})[['Book-Title', 'Book-Author', 'Publisher', 'Avg-Rating']].round(3)
```

**6) For each book in merged books compute the minimum and maximum of the average ratings over all corresponding books in the books dataset.Hence for each book in merged books we will have exactly two values (a minimum and a maximum)**

```python
# Calcoliamo la media di Book-Rating per ogni libro all'interno della tupla (Autore-Titolo-Editore)  
isbn_avg_rating = merged_books_ratings.groupby(['Book-Title', 'Book-Author', 'Publisher', 'ISBN'], as_index = False).mean('Book-Rating')
max_avg = isbn_avg_rating.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index = False).max('Book-Rating').rename(columns={'Book-Rating': 'Max'})[['Book-Title', 'Book-Author', 'Publisher', 'Max']].round(3)
min_avg = isbn_avg_rating.groupby(['Book-Title', 'Book-Author', 'Publisher'], as_index = False).min('Book-Rating').rename(columns={'Book-Rating': 'Min'})[['Book-Title', 'Book-Author', 'Publisher', 'Min']].round(3)
max_min_ratings_avg = pd.merge(min_avg, max_avg, on = ['Book-Title', 'Book-Author', 'Publisher'])
max_min_ratings_avg
```

**7) For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text.We will call this quantity the shared number of reviews with a text.**


```python
def count_element():
    authors_reviews = goodbooks.groupby('authors', as_index = False).sum('work_text_reviews_count').rename(columns = {'work_text_reviews_count':
                    'shared number of reviews with a text'})[['authors','book_id','shared number of reviews with a text', 'original_publication_year']]
    split_author = []
    #Per ogni riga di authors_reviews assegniamo delle variabili
    for axis in authors_reviews.values:     
        reviews = axis[2]
        original_publication_year = axis[3]
        author = axis[0].split(', ')
        #Attribuiamo a ogni autore le reviews equamente divise
        for i in range(len(author)):
            split_author.append([author[i], int(reviews)/len(author), original_publication_year])
    
    return (split_author)

reviews_autori = pd.DataFrame(count_element(), columns = ['Authors','shared number of reviews with a text', 'original_publication_year'])
total_reviews_for_author = reviews_autori.groupby('Authors', as_index = False).sum('shared number of reviews with a text')
total_reviews_for_author
```

**8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.**

```python
author_year = reviews_autori.groupby(['Authors', 'original_publication_year'], as_index = False).sum('shared number of reviews with a text')
#Estraiamo il valore max di reviews per anno
max_rewies_year = author_year.groupby('original_publication_year', as_index = False).max('shared number of reviews with a text')
best_of_year = pd.merge(author_year, max_rewies_year, on = ['original_publication_year', 'shared number of reviews with a text'])

# Il Min original_publication_year Ã¨ -720 che Ã¨ probabile sia un NaN value. Il max Ã¨ 12010. Escludiamo quei valori di 
# original_publication_year che non hanno senso.
best_of_year = best_of_year[(best_of_year['original_publication_year'] < 2100) & (best_of_year['original_publication_year'] > 0)]
best_of_year.sort_values(by = 'original_publication_year')[['Authors', 'original_publication_year', 'shared number of reviews with a text']]
```

**10. Split the users dataset according to the age. One dataset contains the users with unknown age, one with age 0-14, one with age 15-24, one with age 25-34, and so on.**

```python
users = users.sort_values('Age')
k = []
#Aggiungiamo alla lista i valori nulli e i valori tra 0_14 anni
k.append(users[pd.isnull(users['Age'])])
age_0_14 = users.loc[(users['Age'] >= 0) & (users['Age'] <= 14)]
k.append(age_0_14)
user_age = {}
user_age[0] = k[0]
user_age[1] = k[1]
#Rinomiamo gli elementi del dizionario per velocizzare le query
user_age['NaN'] = user_age[0]
user_age['0_14'] = user_age[1]

#Creiamo el elementi del dizionario
for el in range(2,24):   # (244-14)/10 = 23
    for i in range(14,245,10):
        d = (users[(users['Age'] > (i)) & (users['Age'] <= (i+10))])
        k.append(d)
    #Rinominiamo ciascun elemento del dizionario con il range corrispondente
    user_age[el] = k[el]
    first = str(int(user_age[el]['Age'].min()))
    last = str(int(user_age[el]['Age'].max()))
    user_age[('{}_{}').format(first,last)] = user_age[el]
```

**13.Find the author with the highest average rating according to the goodbooks datasets.Â¶**

```python
def split():
    split_author = []
    for axis in goodbooks[['authors', 'average_rating']].values:
        average_rating = axis[1]
        a = axis[0].split(', ')
        #Per ogni riga attribuiamo ad ogni autore lo stesso average_rating
        for el in range(len(a)):
            split_author.append([a[el], average_rating])
    avg_rating_author = pd.DataFrame(data = split_author, columns=['authors', 'avg_rating'])
    avg_author = avg_rating_author.groupby('authors', as_index = False).mean().sort_values(by = 'avg_rating')
    best_author = avg_author.iloc[[-1]]
    return (best_author)
split()
```
